```python
import os
import numpy as np
import pandas as pd
import algo_lib
pd.options.display.max_columns = 150
pd.options.display.max_rows = 3000
pd.options.display.width = 8000
pd.options.mode.chained_assignment = None

stock_i = 'HK.800000'

dataset = pd.read_csv('Database/' + stock_i + '/' + stock_i + '_min.csv', index_col = 0)
dataset.reset_index(inplace=True, drop=True)
#dataset.drop(['turnover_rate', 'volume', 'pe_ratio', 'change_rate'], axis=1, inplace=True)
dataset['high low dif'] = dataset['high'] - dataset['low']
dataset['open close dif'] = dataset['open'] - dataset['close']

dropList = []
for i in range(1, 31):      #比較N分鐘前的變動
    dataset['p%sb' % (i)] = dataset['close'].shift(i)
    dataset['ch%sb' % (i)] = dataset['close'] - dataset['p%sb' % (i)]
    dataset['high%sb' % (i)] = dataset['high'].shift(i)
    dataset['high ch%sb' % (i)] = dataset['high'] - dataset['high%sb' % (i)]
    dataset['low%sb' % (i)] = dataset['low'].shift(i)
    dataset['low ch%sb' % (i)] = dataset['low'] - dataset['low%sb' % (i)]
    dropList.append('p%sb' % (i))
    dropList.append('high%sb' % (i))
    dropList.append('low%sb' % (i))

for j in range(10, 100, 10):    #成交增長
    dataset['%sbuying power' %(j)] = dataset['turnover'] - dataset['turnover'].rolling(j).mean()

for k in range(1, 6):
    minCount = k * 330
    dataset['%sday high' % (k)] = dataset['high'].rolling(minCount).max()
    dataset['%s high dif' % (k)] = dataset['%sday high' % (k)] - dataset['close']
    dataset['%sday low' % (k)] = dataset['low'].rolling(minCount).min()
    dataset['%s low dif' % (k)] = dataset['%sday low' % (k)] - dataset['close']
    dropList.append('%sday high' % (k))
    dropList.append('%sday low' % (k))

for m in range(1, 2):      #下一分鐘的變動
    dataset['next %s high' %(m)] = dataset['high'].shift(-m)
    dataset['next %s high ch' %(m)] = dataset['next %s high' %(m)] - dataset['high']
    dataset['next %s low' %(m)] = dataset['low'].shift(-m)
    dataset['next %s low ch' %(m)] = dataset['next %s low' %(m)] - dataset['low']
    dataset['next %s close' %(m)] = dataset['close'].shift(-m)
    dataset['next %s ch' %(m)] = dataset['next %s close' %(m)] - dataset['close']
    dropList.append('next %s high' %(m))
    dropList.append('next %s low' %(m))
    dropList.append('next %s close' %(m))

for n in range(10, 100, 10):
    dataset['%s std' %(n)] = dataset['close'].rolling(n).std()

'''
dataset['date'] = [x[:10] for x in dataset['time_key']]
dateList = dataset['date'].tolist()
dateList = list(set(dateList))
dateList.sort()

dataset['opening time'] = None
for o in range(len(dateList)):
    OT = dataset.loc[dataset['date'] == dateList[o]].index.tolist()
    openT = 0
    for p in OT:
        dataset['opening time'][p] = openT
        openT += 1

dataset['day high'] = None
dataset['day low'] = None
for ii in range(len(dataset)):
    opentime = dataset['opening time'][ii]
    dataset['day high'][ii] = dataset['high'][ii - opentime:ii + 1].max()
    dataset['day low'][ii] = dataset['low'][ii - opentime:ii + 1].min()

dataset['day high dif'] = dataset['day high'] - dataset['mid']
dataset['day high dif'] = pd.to_numeric(dataset['day high dif'],errors = 'coerce')
dataset['day low dif'] = dataset['day low'] - dataset['mid']
dataset['day low dif'] = pd.to_numeric(dataset['day high dif'],errors = 'coerce')
'''
dataset.drop(dropList, axis = 1, inplace = True)
#'opening time', 'day high', 'day low', 'date'
dataset.drop(['open', 'high', 'low', 'turnover', 'time_key', 'close', 'pe_ratio', 'turnover_rate', 'volume', 'change_rate'], axis = 1, inplace = True)
dataset.drop(range(20000), inplace = True)
dataset.reset_index(inplace=True, drop=True)
dataset.drop(len(dataset)-1, inplace = True)

#2021年6月28日早上停市
print(dataset.head(50))
print(dataset.tail(50))
```

```python
data_num = dataset.shape[0]
train_data = dataset.loc[:data_num * 0.8]
test_data = dataset.loc[data_num * 0.8:]

#標準化
mean = train_data.mean()
std = train_data.std()
train_data = (train_data - mean) / std
test_data = (test_data - mean) / std

input_train = np.array(train_data.drop(['next 1 high ch', 'next 1 low ch', 'next 1 ch'], axis = 1))
target_train = np.array(train_data[['next 1 high ch', 'next 1 low ch']])

input_test = np.array(test_data.drop(['next 1 high ch', 'next 1 low ch', 'next 1 ch'], axis = 1))
target_test = np.array(test_data[['next 1 high ch', 'next 1 low ch']])

print(input_test[:3])
print(input_test.shape)
```

```python
import tensorflow as tf
from tensorflow import keras

model = keras.models.Sequential(name = 'model-%s' %(stock_i),
                                layers = [keras.layers.Flatten(input_shape = (len(train_data.columns) - 3, )),
                                          keras.layers.Dense(128, activation = 'relu'),
                                          keras.layers.Dense(64, activation = 'relu'),
                                          keras.layers.Dense(32, activation = 'relu'),
                                          keras.layers.Dense(2)])

model.summary()

model.compile(optimizer = keras.optimizers.Adam(0.001),
              loss = keras.losses.Huber(delta=1.0),
              metrics = [keras.metrics.MeanAbsoluteError()])

model_dir = 'Tensor/models/'
log_dir = os.path.join('Tensor', 'model-%s' %(stock_i))
model_cbk = tf.keras.callbacks.TensorBoard(log_dir = log_dir)
model_mckp = tf.keras.callbacks.ModelCheckpoint('Tensor/model-%s/Best-model8-%s.h5' %(stock_i, stock_i),
                                             monitor = 'mean_absolute_error', save_best_only = True, mode = 'min')
history = model.fit(input_train, target_train, batch_size = 256, epochs = 300, callbacks = [model_cbk, model_mckp])
```
